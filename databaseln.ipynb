```python
import sqlite3

# 1. Establish the 'Bridge'
conn = sqlite3.connect('practice.db')

# 2. Get the 'Waiter/Courier' (The Cursor)
cursor = conn.cursor()

# 3. Use the cursor to CREATE a table
# (We use IF NOT EXISTS so it doesn't error if you run it twice)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS employees (
        name TEXT,
        age INTEGER,
        salary REAL
    )
''')

# 4. Use the cursor to INSERT data
# Data Scientists often use lists of tuples like this:
staff = [('Alice', 25, 1500), ('Bob', 30, 2200), ('Charlie', 40, 3500)]
cursor.executemany('INSERT INTO employees VALUES (?,?,?)', staff)

# 5. COMMIT the changes to the database file
conn.commit()

# 6. Use the cursor to FETCH data back
print("--- Employee List ---")
cursor.execute("SELECT * FROM employees")

# We loop through the cursor object directly
for row in cursor:
    print(f"Name: {row[0]}, Age: {row[1]}, Salary: ${row[2]}")

# 7. Close the bridge
conn.close()
```

```python
import sqlite3

class HospitalDatabase:
    def __init__(self, cursor):
        """Initialize with a database cursor and create tables if they don't exist."""
        self.cursor = cursor
        self.create_tables()

    def create_tables(self):
        """Creates the necessary tables for patients and appointments."""
        # Create Patients Table
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS patients (
                patient_id INTEGER PRIMARY KEY AUTOINCREMENT,
                f_name TEXT, l_name TEXT, dob TEXT, gender TEXT, 
                blood_type TEXT, phone TEXT, email TEXT, address TEXT, 
                emer_contact TEXT
            )
        """)
        
        # Create Appointments Table
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS appointments (
                app_id INTEGER PRIMARY KEY AUTOINCREMENT,
                patient_id INTEGER, doctor_id INTEGER,
                date TEXT, time TEXT, reason TEXT,
                FOREIGN KEY(patient_id) REFERENCES patients(patient_id)
            )
        """)

    def add_patient(self, f_name, l_name, dob, gender, blood, phone, email, addr, emer):
        """Inserts a new patient into the database."""
        query = """
            INSERT INTO patients 
            (f_name, l_name, dob, gender, blood_type, phone, email, address, emer_contact) 
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
        """
        self.cursor.execute(query, (f_name, l_name, dob, gender, blood, phone, email, addr, emer))
        print(f"Patient {f_name} {l_name} added successfully.")

    def schedule_appointment(self, patient_id, doctor_id, date, time, reason):
        """Schedules an appointment."""
        query = """
            INSERT INTO appointments (patient_id, doctor_id, date, time, reason)
            VALUES (?, ?, ?, ?, ?)
        """
        self.cursor.execute(query, (patient_id, doctor_id, date, time, reason))
        print("Appointment scheduled.")

    def search_patients(self, name_keyword):
        """Search for a patient by first name."""
        query = "SELECT * FROM patients WHERE f_name LIKE ?"
        self.cursor.execute(query, ('%' + name_keyword + '%',))
        return self.cursor.fetchall()

    def get_statistics(self):
        """Returns a count of total patients."""
        self.cursor.execute("SELECT COUNT(*) FROM patients")
        count = self.cursor.fetchone()[0]
        return {"total_patients": count}

# --- MAIN EXECUTION BLOCK ---

# Open the connection (this creates the file 'HospitalDatabase.db' if it doesn't exist)
with sqlite3.connect('HospitalDatabase.db') as conn:
    cursor = conn.cursor()
    
    # 1. Instantiate the class (This triggers __init__ and creates tables)
    hospital = HospitalDatabase(cursor)

    # 2. Add patient
    hospital.add_patient('John', 'Doe', '1985-05-15', 'Male', 'O+', 
                        '555-1234', 'john@email.com', '123 Main St', '555-9999')

    # 3. Schedule appointment
    hospital.schedule_appointment(patient_id=1, doctor_id=1, 
                                 date='2024-12-25', time='10:00 AM', 
                                 reason='Annual checkup')

    # 4. Search patients
    results = hospital.search_patients('John')
    print(f"\nSearch Results for 'John':\n{results}")

    # 5. Get statistics
    stats = hospital.get_statistics()
    print(f"\nHospital Stats: {stats}")
    
    # 6. Save changes
    conn.commit()
```

Data Types in SQLite:

INTEGER - Whole numbers
REAL - Decimal numbers
TEXT - Strings
BLOB - Binary data
NULL - No value

Constraints:

PRIMARY KEY - Unique identifier
AUTOINCREMENT - Auto-generate ID
NOT NULL - Must have value
UNIQUE - No duplicates
DEFAULT - Default value
CHECK - Validation rule

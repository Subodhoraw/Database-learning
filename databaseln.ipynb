```python
import sqlite3

# 1. Establish the 'Bridge'
conn = sqlite3.connect('practice.db')

# 2. Get the 'Waiter/Courier' (The Cursor)
cursor = conn.cursor()

# 3. Use the cursor to CREATE a table
# (We use IF NOT EXISTS so it doesn't error if you run it twice)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS employees (
        name TEXT,
        age INTEGER,
        salary REAL
    )
''')

# 4. Use the cursor to INSERT data
# Data Scientists often use lists of tuples like this:
staff = [('Alice', 25, 1500), ('Bob', 30, 2200), ('Charlie', 40, 3500)]
cursor.executemany('INSERT INTO employees VALUES (?,?,?)', staff)

# 5. COMMIT the changes to the database file
conn.commit()

# 6. Use the cursor to FETCH data back
print("--- Employee List ---")
cursor.execute("SELECT * FROM employees")

# We loop through the cursor object directly
for row in cursor:
    print(f"Name: {row[0]}, Age: {row[1]}, Salary: ${row[2]}")

# 7. Close the bridge
conn.close()
```

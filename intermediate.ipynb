### 2.1 Filtering and sorting 

```python
import sqlite3
conn = sqlite3.connect('library.db')
cursor = conn.cursor()

#WHERE with multiple conditions
cursor.execute(''' SELECT * FROM patients WHERE gender = ? AND date_of_birth > ? ''',('Female','1985-01-01'))

#OR CONDITION 
cursor.execute('''SELECT * FROM patients WHERE gender = ? OR date_of_birth < ? ''',('Male','1980-01-01'))

#LIKE pattern matching 
cursor.execute("SELECT * FROM patients WHERE last_name LIKE 'S%'") #STARTS WITH S
cursor.execute(" SELECT * FROM patients WHERE email LIKE '@gmail.com'") # gmail user

#In operator
cursor.execute('''SELECT * FROM patients
               WHERE gender IN ('Male','Female')''')

#Between 
cursor.execute('''SELECT *FROM patients
               WHERE date_of_birth BETWEEN ? AND ? ''',('1980-01-01','1990-12-31'))

#ORDER BY
cursor.execute('SELECT * FROM patients ORDER BY last_name ASC') # ASCENDING
cursor.execute('SELECT * FROM patients ORDER BY date_of_birth DESC') #Descending

#LIMIT
cursor.execute('SELECT * FROM patients LIMIT 10') #FIRST 10
cursor.execute('SELECT * FROM patients LIMIT 10 OFFSET 20') #skip 20 get 10
conn.close()
```

### adding new table 

```python
conn = sqlite3.connect('Sakila.db')
cursor = conn.cursor()
cursor.execute(''' CREATE TABLE film_reviwes(
               review_id INT PRIMARY KEY AUTOINCREMENT,
               film_id INTEGER NOT NULL,
               customer_id INTEGER NOT NULL,
               review_text TEXT,
               rating INTEGER CHECK (rating BETWEEEN 1 AND 5),
               review_date DATETIME DEFAULT CURRENT_TIMESTAMP,
               ----linking to existing SAKILA TABLES
               FOREIGN KEY (film_id) REFERENCES film(film_id),
               FOREIGN KEY (customer_id) REFERENCES customer(customer_id)
               )''')
conn.commit()
conn.close()
```
